Based on the information above, here are the changes we will make to the datasets.

1.   Process social_df column names


2.   Remove rows with NaN values
*   mental_df['severity']
*   social_df['org_affiliation']


3.   Remove rows with undefinable responses
*   social_df['gender'] = ['unsure', 'Trans', 'There are others???']


4.   Change data types
*   social_df['timestamp'] > extract data as DateTime
*   social_df['age'] > convert to int to match other dataset

5.  Clean social_df data
*   social_df['gender'] > convert all non-binary options into a single option
*   social_df['org_affiliation'] > create dummy columns for University, Private, School, Company, Government (aka "Goverment")
*   social_df['most_used_platforms] > create dummy columns for Facebook, Instagram, YouTube, Snapchat, Discord, Reddit, Pinterest, TikTok, Twitter







```python
# 1. Process social_df column names
social_initial_cols = social_df_raw.columns
print(social_initial_cols)

# Define new column names
clean_cols = [
    'timestamp',
    'age',
    'gender',
    'relationship_status',
    'occupation_status',
    'org_affiliation',
    'sm_usage', # sm = social media
    'sm_platforms',
    'sm_time_daily',
    'sm_without_purpose_freq',
    'sm_distraction_freq',
    'sm_restless_if_unused',
    'distractibility',
    'worried',
    'concentration_difficulty',
    'sm_comparison_freq',
    'sm_comparison_feelings',
    'sm_validation_seeking_freq',
    'depressed_freq',
    'interest_fluctuation',
    'sleep_issues'
]

#Create a clean version of the dataframe
social_df_clean = social_df_raw.copy() # Create copy for processed df
social_df_clean.columns = clean_cols
social_df_clean.head(1)
```

```python
# 2. Remove rows with NaN values
mental_df_clean = mental_df_raw.copy() # Create copy for processed df

mental_df_clean = mental_df_clean.dropna(subset=['Severity'])
social_df_clean = social_df_clean.dropna(subset=['org_affiliation'])

# Check
print(mental_df_clean.isna().sum())
print(social_df_clean.isna().sum())
```

```python
# 3. Remove rows with undefinable responses
social_df_clean = social_df_clean[~social_df_clean['gender'].isin(['unsure ', 'Trans', 'There are others???'])]
print(social_df_clean['gender'].unique())
```

```python
# 4. Convert datatypes as needed
social_df_clean['timestamp'] = pd.to_datetime(social_df_clean['timestamp'], errors='coerce') # convert timestamp to datetime
social_df_clean['age'] = np.floor(social_df_clean['age']).astype('Int64') # convert age to int (to match mental_df)
social_df_clean.head()
```

```python
# 5. Clean data in social_df_clean (mental_df is already clean)

# Convert all non-binary options into a single option
print(social_df_clean['gender'].value_counts())
nb_alias = ['Nonbinary ', 'Non-binary', 'Non binary ']
social_df_clean['gender'] = social_df_clean['gender'].apply(
    lambda x: 'Non-binary' if x in nb_alias else x
)
print(social_df_clean['gender'].unique())

# Create dummy columns for University, Private, School, Company, Government (aka "Goverment")
social_df_clean['org_affiliation'] = social_df_clean['org_affiliation'].astype(str) # cast column values into string
explode_org = social_df_clean.assign(org=social_df_clean['org_affiliation'].str.split(',')).explode('org') # explode each row into multiple rows per org listed
explode_org['org'] = explode_org['org'].str.strip() # strip whitespace
org_dummies = pd.get_dummies(explode_org['org']) # create dummies
grouped_org_dummies = org_dummies.groupby(explode_org.index).max() # group dummies by orignal index
social_df_clean = pd.concat([social_df_clean, grouped_org_dummies], axis=1) # join with original dataframe
social_df_clean = social_df_clean.drop('N/A', axis=1) # drop the N/A affiliation column
social_df_clean = social_df_clean.rename(columns={'Goverment': 'Government'}) # rename Goverment to Government

# Create dummy columns for Facebook, Instagram, YouTube, Snapchat, Discord, Reddit, Pinterest, TikTok, Twitter
social_df_clean['sm_platforms'] = social_df_clean['sm_platforms'].astype(str) # cast column values into string
explode_plat = social_df_clean.assign(platform=social_df_clean['sm_platforms'].str.split(',')).explode('platform') # explode each row into multiple rows per org listed
explode_plat['platform'] = explode_plat['platform'].str.strip() # strip whitespace
plat_dummies = pd.get_dummies(explode_plat['platform']) # create dummies
grouped_plat_dummies = plat_dummies.groupby(explode_plat.index).max() # group dummies by orignal index
social_df_clean = pd.concat([social_df_clean, grouped_plat_dummies], axis=1) # join with original dataframe

social_df_clean.head()
```

```python
print(f'Final shape of mental health dataframe: {mental_df_clean.shape}') # (50000, 17) > (24998,17)
print(f'Final shape of mental health dataframe: {social_df_clean.shape}') # (481, 21) > (449, 35)
```
